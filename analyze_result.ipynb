```python
import soundfile as sf
import numpy as np
import matplotlib.pyplot as plt
y1,sr1 = sf.read('data\RIRs\s1_p01_m1.wav')
y2,sr2 = sf.read('data\RIRs\s1_p01i_m1.wav')
y3,sr3 = sf.read('data\RIRs\s1_p01r_m1.wav')
time1 = np.arange(0, len(y1)) / sr1
time2 = np.arange(0, len(y2)) / sr2
time3 = np.arange(0, len(y3)) / sr3
plt.plot(time1, y1)
# plt.plot(time2, y2)
plt.plot(time3, y3)
plt.xlim(1, 3)
```
